## Segment 2 - Defining elements of a plot

```python
import numpy as np
from numpy.random import randn
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from matplotlib import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = (10, 8)
```

### Defining axes, ticks, and grids

```python
x = range(1,10)
y = [1,2,3,4,0,4,3,2,1]

fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])
ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.plot(x,y)
```

```python
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

ax.set_xlim([1,9])
ax.set_ylim([0,5])

ax.set_xticks([0,1,2,4,5,6,8,9,10])
ax.set_yticks([0,1,2,3,4,5])

ax.grid()
ax.plot(x,y)
```

### Generating multiple plots in one figure with subplots 

```python
fig = plt.figure()
fig, (ax1, ax2) = plt.subplots(1, 2)

ax1.plot(y,x)
ax2.plot(x,y)
```

## Segment 4 - Creating labels and annotations

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

import matplotlib.pyplot as plt
from pylab import rcParams
```

```python
%matplotlib inline
rcParams['figure.figsize'] = 8,4
```

### Labeling plot features
#### The functional method

```python
x = range(1,10)
y = [1,2,3,4,.5,4,3,2,1]
plt.bar(x, y)
plt.xlabel('your x-axis label')
plt.ylabel('your y-axis label')
```

```python
z = [1,2,3,4,.5]
veh_type = ['bicycle','motor bike', 'stroller', 'van', 'bus']

plt.pie(z, labels= veh_type)
plt.show()
```

#### The object-oriented method

```python
address = "C:\\Users\\Mattbot\\Documents\\Python Data Science\\Ex_Files_Python_Data_Science_EssT_Pt_1\\Exercise Files\\Data\\mtcars.csv"

cars = pd.read_csv(address)
cars.columns = ['car_names','mpg','cyl', 'disp', 'hp', 'drat', 'wt','qsec', 'vs', 'am', 'gear', 'carb']
cars.head()

mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 45, fontsize = 'medium')
ax.set_title('Mpg of cars in mtcars data set')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')
ax.grid()
```

### Adding a legend to your plot
#### The functional method

```python
plt.pie(z)
plt.legend(veh_type, loc = 'best')
plt.show()
```

#### The object-oriented method

```python
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 45, fontsize = 'medium')
ax.set_title('Mpg of cars in mtcars data set')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')
ax.legend(loc = 'best')

ax.grid()
```

### Annotating your plot

```python
mpg.max()
```

```python
mpg = cars.mpg
fig = plt.figure()
ax = fig.add_axes([.1,.1,1,1])

mpg.plot()

ax.set_xticks(range(32))

ax.set_xticklabels(cars.car_names, rotation = 45, fontsize = 'medium')
ax.set_title('Mpg of cars in mtcars data set')

ax.set_xlabel('Car Names')
ax.set_ylabel('Miles Per Gallon')
ax.legend(loc = 'best')

ax.set_ylim([0,45])
ax.annotate('Toyota Corolla', xy=(19,33.9), xytext = (21,35),
           arrowprops= dict(facecolor = 'black', shrink=0.05))

ax.grid()
```

## Segment 6 - Creating statistical data graphics

```python
import numpy as np
import pandas as pd
from pandas import Series, DataFrame

from pandas.plotting import scatter_matrix

import matplotlib.pyplot as plt
from pylab import rcParams

%matplotlib inline
rcParams['figure.figsize'] = 10, 8

import seaborn as sb
sb.set_style('whitegrid')
```

### Eyeballing dataset distributions with histograms

```python
address = "C:\\Users\\Mattbot\\Documents\\Python Data Science\\Ex_Files_Python_Data_Science_EssT_Pt_1\\Exercise Files\\Data\\mtcars.csv"

cars = pd.read_csv(address)

cars.columns = ['car_names','mpg','cyl','disp', 'hp', 'drat', 'wt', 'qsec', 'vs', 'am', 'gear', 'carb']

cars.index = cars.car_names

mpg = cars['mpg']

mpg.plot(kind='hist')
```

```python
plt.hist(mpg)
plt.plot()
```

```python
sb.distplot(mpg)
```

###  Seeing scatterplots in action

```python
cars.plot(kind = 'scatter', x = 'hp', y= 'mpg', c='darkgreen', s=150)
```

```python
sb.regplot(x = 'hp', y ='mpg', data=cars, scatter=True, color='darkgreen')
```

### Generating a scatter plot matrix

```python
sb.pairplot(cars)
```

```python
cars_subset = cars[['mpg', 'disp', 'hp', 'wt']]
sb.pairplot(cars_subset)
plt.show()
```

### Building boxplots

```python
cars.boxplot(column='mpg', by='am')
cars.boxplot(column='wt', by='am')
```

```python
sb.boxplot(x='am', y='mpg', data=cars, palette='hls')
```
